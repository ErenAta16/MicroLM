# 01 — Pretrain Ladder

Pretrain each rung from `configs/model_ladder.yaml`.

```python
DRIVE = '/content/drive/MyDrive/hollow-chains'
REPO = '/content/hollow-chains'
%cd $REPO
```

```python
from hollow_chains.config import load_config
from hollow_chains.train.pretrain import run_pretrain
from hollow_chains.utils.seed import set_seed

cfg = load_config('configs/pretrain.yaml')
cfg['paths']['checkpoint_dir'] = f'{DRIVE}/checkpoints/pretrain'
cfg['tokenizer']['path'] = f'{DRIVE}/artifacts/tokenizer'
cfg['data']['shard_dir'] = f'{DRIVE}/artifacts/pretrain_shards'
set_seed(cfg['seed'])
for rung in ['tiny_1m','small_8m','mid_50m','large_150m']:
    cfg['training']['rung'] = rung
    ckpt = run_pretrain(cfg)
    print(rung, '->', ckpt)
```
