# 02 — SFT Sweeps

Build SFT data and run one-axis sweep cells from `configs/sft.yaml`.

```python
DRIVE = '/content/drive/MyDrive/hollow-chains'
%cd /content/hollow-chains
```

```python
from hollow_chains.config import load_config
from hollow_chains.data.build_reasoning_sft import build_sft_jsonl
from hollow_chains.train.sft import run_sft

cfg = load_config('configs/sft.yaml')
cfg['paths']['pretrain_checkpoint_root'] = f'{DRIVE}/checkpoints/pretrain'
cfg['paths']['sft_checkpoint_root'] = f'{DRIVE}/checkpoints/sft'
for cell in cfg['run_cells']:
    sft_path, rate = build_sft_jsonl(cfg, teacher=cell['teacher'], samples=cell['samples'], fmt=cell['format'])
    pretrain_ckpt = f"{DRIVE}/checkpoints/pretrain/{cell['rung']}/final"
    run_sft(cfg, rung=cell['rung'], teacher=cell['teacher'], samples=cell['samples'], epochs=cell['epochs'], fmt=cell['format'], pretrain_checkpoint=pretrain_ckpt, sft_jsonl=sft_path)
```
