```python
from hollow_chains.config import load_config
from hollow_chains.data.tokenizer import train_tokenizer
from hollow_chains.data.pretrain_data import get_corpus_texts, materialize_shard
from hollow_chains.data.tokenizer import load_tokenizer
from hollow_chains.utils.seed import set_seed

set_seed(SEED)
tok_cfg = load_config('configs/tokenizer.yaml')
pre_cfg = load_config('configs/pretrain.yaml')
texts = get_corpus_texts(tok_cfg)
train_tokenizer(texts, f'{DRIVE}/artifacts/tokenizer', vocab_size=tok_cfg['vocab_size'], config=tok_cfg)
tok = load_tokenizer(f'{DRIVE}/artifacts/tokenizer')
materialize_shard(texts[:50000], tok, f'{DRIVE}/artifacts/pretrain_shards', context_len=512)
print('Artifacts:', f'{DRIVE}/artifacts/tokenizer', f'{DRIVE}/artifacts/pretrain_shards')
```
